```python
# Custom functions to manage SHU Dataset
def load_pretraining(path, eegsym=True):
    EEG = loadmat(path, simplify_cells=True)['DATA_STRUCT']['data']
    chan_reorder = [0, 31, 35, 4, 37, 6, 39, 10, 43, 
                    8, 41, 45, 12, 47, 14, 49, 18, 
                    53, 16, 51, 28, 20, 55, 22, 57, 
                    24, 59, 26, 25, 60]
    EEG = EEG[chan_reorder,:]
    if eegsym:
        sym_reorder=[0, 3, 5, 7, 9, 12, 14, 16, 18,
                     21, 23, 25, 28, 2, 11, 20, 27,
                     29, 29, 24, 22, 19, 17, 15, 13,
                     10, 8, 6, 4, 1]
        EEG = EEG[sym_reorder,:]
    return EEG

def load_SHU_EEG(path, eegsym=True):
    data = loadmat(path, simplify_cells=True)['DATA_STRUCT']
    EEG = data['data']
    EEG = np.delete(EEG, (16,17), axis=0)
    if eegsym:
        sym_reorder=[0, 3, 5, 7, 9, 12, 14, 16, 18,
                     21, 23, 25, 28, 2, 11, 20, 27,
                     29, 29, 24, 22, 19, 17, 15, 13,
                     10, 8, 6, 4, 1]
        EEG = EEG[sym_reorder,:]
    label = data['label']-1
    return EEG, label

def transform_SHU_EEG(EEG, Range=200, asintote=2, scale='uV', exact=True):
    return utils.scale_range_soft_clip(EEG, Range, asintote, scale, exact)
```

## Define and Visualize Data Augmentation Module

```python
# DEFINE AUGMENTER
# FIRST RANDOM SELECTION: APPLY FLIP OR CHANGE REFERENCE OR NOTHING
AUG_ref = aug.StaticSingleAug(aug.change_ref, {'reference': 45})
AUG_flipv = aug.StaticSingleAug(aug.flip_vertical)
AUG_flipr = aug.StaticSingleAug(aug.flip_horizontal)
AUG_id = aug.StaticSingleAug(aug.identity)
Sequence1 = aug.RandomAug( AUG_ref, AUG_flipv, AUG_id)
Sequence1 = aug.RandomAug( AUG_id)

# SECOND RANDOM SELECTION: ADD SOME NOISE
AUG_bandN = aug.DynamicSingleAug(aug.add_band_noise, 
                                 discrete_arg={'bandwidth': ["delta", "theta", "alpha", "beta", (30,49) ], 
                                               'samplerate': [freq],
                                               'noise_range': 0.005
                                              }
                                )
AUG_artifact = aug.StaticSingleAug(aug.add_eeg_artifact, {'Fs': freq, 'amplitude':0.005})
Sequence2 = aug.RandomAug( AUG_bandN, AUG_artifact)

# THIRD RANDOM SELECTION: CROP OR RANDOM PERMUTATION
AUG_crop = aug.DynamicSingleAug(aug.crop_and_resize,
                                discrete_arg={'batch_equal': False},
                                range_arg ={'N_cut': [1, 4], 'segments': [10,15]},
                                range_type ={'N_cut': True, 'segments': True})

AUG_permute = aug.DynamicSingleAug( aug.permutation_signal,
                                    discrete_arg={'segments': [10], 'batch_equal': False},
                                    range_arg ={'seg_to_per': [2, 5]},
                                    range_type={'seg_to_per': True}
                                  )
Sequence3 = aug.RandomAug( AUG_crop, AUG_permute)

# FINAL AUGMENTER: SEQUENCE OF THE THREE RANDOM LISTS
Augmenter = aug.SequentialAug(Sequence1,Sequence2,Sequence3)
```

## Define Encoder, SSL Architecture and other training modules

```python
#DEFINE ENCODER ARCHITECTURE
factor = 2
kernLength = 64*factor
dropRate = 0.2
F1 = 8
D = 2
F2 = 16
ELUalpha = 1
dropType = 'Dropout'
pool1=4*factor
pool2=8*factor
separable_kernel=16*factor
depthwise_max_norm=1.

NNencoder= zoo.EEGNetEncoder(Chans=chans, kernLength = kernLength, dropRate = dropRate, 
                             F1 = F1, D = D, F2 = F2, ELUalpha= ELUalpha, dropType = dropType,
                             pool1 = pool1, pool2 = pool2, separable_kernel = separable_kernel, 
                             depthwise_max_norm= depthwise_max_norm
                            )
NNencoder2= copy.deepcopy(NNencoder) # Copy of the random initialization for embedding analysis

# DEFINE SSL ARCHITECTURE
emb_size= F2*((freq*window)//int(pool1*pool2))
head_size=[ emb_size, 128, 64]
predictor_size=[head_size[-1],head_size[-1]]

SelfMdl = ssl.SimCLR( encoder=NNencoder, projection_head=head_size).to(device=device)
loss=losses.SimCLR_loss
loss_arg={'temperature': 0.5}

#SelfMdl = ssl.SimSiam( encoder=NNencoder, projection_head=head_size, 
#                       predictor=predictor_size).to(device=device)
#loss=losses.SimSiam_loss
#loss_arg=None

# SelfMdl = ssl.MoCo(encoder=NNencoder, 
#                    projection_head=[256,128,128,256], 
#                    predictor=[256,128,256],
#                    bank_size=0 #1024
#                   ).to(device=device)

#SelfMdl = ssl.BYOL( encoder=NNencoder, projection_head=head_size, 
#                      predictor=predictor_size).to(device=device)
#loss=losses.BYOL_loss
#loss_arg=[]

# SelfMdl = ssl.Barlow_Twins( encoder=NNencoder, projection_head=[256,128,128]).to(device=device)


# DEFINE EARLYSTOPPER
earlystop = ssl.EarlyStopping(patience=150, min_delta=1e-05, record_best_weights=True)

# DEFINE OPTIMIZER 
optimizer = torch.optim.Adam(SelfMdl.parameters(), lr=1e-3)

# DEFINE SCHEDULER
scheduler = torch.optim.lr_scheduler.ExponentialLR(optimizer, gamma=0.97)
```

## Start Pretraining 

```python
# -----------------------
#   SET FIT PARAMETERS
# -----------------------
epochs= 1
start= time.time()
loss_info = []
loss_info = SelfMdl.fit(train_dataloader = trainloader,
                        augmenter=Augmenter,
                        epochs=epochs,
                        optimizer=optimizer,
                        loss_func= loss,
                        loss_args= loss_arg,
                        lr_scheduler= scheduler,
                        EarlyStopper=earlystop,
                        validation_dataloader=valloader,
                        verbose=1,
                        device= device,
                        validation_freq=1,
                        return_loss_info=True
                       )
print('training concluded in: ', (time.time() - start)/(60*60), ' hours')
```

```python
# PLOT LOSS PER EPOCH
is_val=True
train_loss= []
val_loss= []
if loss_info==[]:
    print('No available loss_info')
else:
    for i in loss_info:
        train_loss.append( sum(loss_info[i][0]) / len(loss_info[i][0]))
        if is_val:
            val_loss.append( sum(loss_info[i][1]) / len(loss_info[i][1]))
        
    epochs=[i for i in range(1, len(train_loss)+1)]
    
    plt.plot(epochs, train_loss, 'o-')
    if is_val:
        plt.plot(epochs, val_loss, 'o-')
    plt.title('Loss Progression')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend(['Training loss', 'validation loss'])
    plt.show()
```

### Define Fine-tuning Model and training elements

```python
# DEFINE FINE-TUNING MODEL
nb_classes = 2

factor = 4
kernLength = 64*factor
dropRate = 0.5
F1 = 32
D = 4
F2 = 64
ELUalpha = 1
dropType = 'Dropout'
pool1=4*factor
pool2=8*factor
separable_kernel=16*factor
depthwise_max_norm=2.
Samples = int(freq*window)
FinalMdl = zoo.EEGNet(nb_classes = nb_classes, 
                      Chans = chans, 
                      Samples = Samples, 
                      kernLength = kernLength, 
                      dropRate = dropRate, 
                      F1 = F1, 
                      D = D, 
                      F2 = F2, 
                      norm_rate=2.0, 
                      dropType = dropType, 
                      ELUalpha= ELUalpha, 
                      pool1 = pool1, 
                      pool2 = pool2, 
                      separable_kernel = separable_kernel, 
                      depthwise_max_norm= depthwise_max_norm 
                     )
FinalMdl_Base = copy.deepcopy(FinalMdl)

# Transfer the pretrained backbone
SelfMdl.train()           # To be sure that require grad is set to true on all weights
SelfMdl.to(device='cpu')  # To free some memory on GPU before fine-tuning

if device.type=='cuda':
    torch.cuda.empty_cache()

#FinalMdl.encoder = SelfMdl.get_encoder()

FinalMdl=zoo.ResNet1D(nb_classes= nb_classes, Chans=chans, Samples=Samples, 
                      block=zoo.BasicBlock1, Layers =[2, 2, 2, 2],
                      inplane=32, kernLength=11)

#FinalMdl = zoo.EEGSym(nb_classes, Samples, chans, freq, scales_time=(250, 125, 64),
#                      lateral_chans=13, F = 8, pool=2, dropRate=0.5)

FinalMdl.train()
FinalMdl.to(device=device)

# DEFINE LOSS
def loss_fineTuning(yhat, ytrue):
    ytrue = ytrue + 0.
    yhat = torch.squeeze(yhat)
    return F.binary_cross_entropy_with_logits(yhat, ytrue)

# DEFINE EARLYSTOPPER
earlystopFT = ssl.EarlyStopping(patience=25, min_delta=1e-03, record_best_weights=True)
earlystopFT2 = copy.deepcopy(earlystopFT)

# DEFINE OPTIMIZER 
lr=1e-3
optimizerFT = torch.optim.Adam(FinalMdl.parameters(), lr=lr)

# DEFINE SCHEDULER
gamma=0.99
schedulerFT = torch.optim.lr_scheduler.ExponentialLR(optimizerFT, gamma=gamma)


# DEFINE OPTIMIZER 
optimizerFT2 = torch.optim.Adam(FinalMdl.parameters(), lr=lr)
schedulerFT2 = torch.optim.lr_scheduler.ExponentialLR(optimizerFT2, gamma=gamma)
```
