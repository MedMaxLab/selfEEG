## Define and Visualize Data Augmentation Module

```python
# DEFINE AUGMENTER
# FIRST RANDOM SELECTION: APPLY FLIP OR CHANGE REFERENCE OR NOTHING
AUG_ref = aug.StaticSingleAug(aug.change_ref, {'reference': 45})
AUG_flipv = aug.StaticSingleAug(aug.flip_vertical)
AUG_flipr = aug.StaticSingleAug(aug.flip_horizontal)
AUG_id = aug.StaticSingleAug(aug.identity)
Sequence1 = aug.RandomAug( AUG_ref, AUG_flipv, AUG_id)
Sequence1 = aug.RandomAug( AUG_id)

# SECOND RANDOM SELECTION: ADD SOME NOISE
AUG_bandN = aug.DynamicSingleAug(aug.add_band_noise, 
                                 discrete_arg={'bandwidth': ["delta", "theta", "alpha", "beta", (30,49) ], 
                                               'samplerate': [freq],
                                               'noise_range': 0.005
                                              }
                                )
AUG_artifact = aug.StaticSingleAug(aug.add_eeg_artifact, {'Fs': freq, 'amplitude':0.005})
Sequence2 = aug.RandomAug( AUG_bandN, AUG_artifact)

# THIRD RANDOM SELECTION: CROP OR RANDOM PERMUTATION
AUG_crop = aug.DynamicSingleAug(aug.crop_and_resize,
                                discrete_arg={'batch_equal': False},
                                range_arg ={'N_cut': [1, 4], 'segments': [10,15]},
                                range_type ={'N_cut': True, 'segments': True})

AUG_permute = aug.DynamicSingleAug( aug.permutation_signal,
                                    discrete_arg={'segments': [10], 'batch_equal': False},
                                    range_arg ={'seg_to_per': [2, 5]},
                                    range_type={'seg_to_per': True}
                                  )
Sequence3 = aug.RandomAug( AUG_crop, AUG_permute)

# FINAL AUGMENTER: SEQUENCE OF THE THREE RANDOM LISTS
Augmenter = aug.SequentialAug(Sequence1,Sequence2,Sequence3)
```

# FINE-TUNING PHASE

```python
is_binary_class=False
nb_classes = 1 if is_binary_class else 4
```

```python
# -------------------------------
#  DEF ONE HOT ENCODING FUNCTION
# -------------------------------
# The current problem is a 4 class classification task: healthy, parkinson, alzheimer, dementia EEGs
# Dataset 5    -->    PD: Parkinson         -       CTL: Control
# Dataset 10   -->     A: Alzheimer    F: Dementia    C: Control

def one_hot_bin(label):
    OneHot= torch.zeros(len(label), dtype=torch.long)
    for n, i in enumerate(label):
        if i in ['A', 'F', 'PD']:
            OneHot[n]=1
        else:
            OneHot[n]=0
    return torch.unsqueeze(OneHot,1)


def one_hot_ND(label):
    OneHot= torch.zeros(len(label), 4, dtype=torch.long)
    for n, i in enumerate(label):
        if i == 'A':
            OneHot[n,1]=1
        elif i == 'F':
            OneHot[n,2]=1
        elif i == 'PD':
            OneHot[n,3]=1
        else:
            OneHot[n,0]=1
    return OneHot


def Encoder_ND(label):
    OneHot= torch.zeros(len(label), dtype=torch.long)
    for n, i in enumerate(label):
        if i == 'A':
            OneHot[n]=1
        elif i == 'F':
            OneHot[n]=2
        elif i == 'PD':
            OneHot[n]=3
        else:
            OneHot[n]=0
    return OneHot
```

```python
def count_classes(dataloader, 
                  nclasses: int=2, 
                  class_name: list[str]=None
                 ):
    class_count=[0]*nclasses
    N = range(nclasses)
    if class_name==None:
        class_name=[i for i in N]
    for X , y in dataloader:
        for i in N:
            try: 
                class_count[i]= class_count[i] + sum(1 for x in y if x in class_name[i])
            except:
                class_count[i]= class_count[i] + sum(1 for x in y if x == class_name[i])
    return class_count

if nb_classes==4:
    print('Class Distribution: Healthy | Alzheimer | Dementia | Parkinson')
    cls_name=[['CTL','C'],'A','F','PD']
else:
    print('Class Distribution: Healthy | Unhealthy')
    cls_name=[['CTL','C'],['A','F','PD']]

train_count= count_classes(trainloaderFT, 2 if nb_classes==1 else nb_classes, cls_name )
print(' Training set:   ', train_count)
val_count= count_classes(valloaderFT, 2 if nb_classes==1 else nb_classes, cls_name)
print(' Validation set:   ', val_count)
test_count= count_classes(testloaderFT, 2 if nb_classes==1 else nb_classes, cls_name)
print(' Test set:   ', test_count)
```

# EMBEDDING ANALYSIS

```python
# GET ALL EMBEDDINGS AND LABEL FOR MAP VISUALIZATION
NNencoder2.eval()
SelfMdl.eval()
FinalMdl.eval()
NNencoder2.to(device='cpu')
SelfMdl.to(device='cpu')
FinalMdl.to(device='cpu')

Emb_Random= torch.zeros(len(testset),F2*((freq*window)//32))
Emb_SSL= torch.zeros_like(Emb_Random)
Emb_FT= torch.zeros_like(Emb_Random)
Label_stack= torch.zeros(len(testset), 1 if nb_classes<=2 else nb_classes )

cnt=0
for i, (X, y) in enumerate(testloader):
    sys.stdout.write('\r')
    sys.stdout.write("[%-20s] %d%%" % ('='*int(20*i/len(testloader)), 100*i/len(testloader)))
    sys.stdout.flush()
    
    with torch.no_grad():
        Emb_Random[cnt:cnt+X.shape[0]]=NNencoder2(X)
        Emb_SSL[cnt:cnt+X.shape[0]]=SelfMdl.encoder(X)
        Emb_FT[cnt:cnt+X.shape[0]]=FinalMdl.encoder(X)
        Label_stack[cnt:cnt+X.shape[0]]=one_hot_ND(y) if nb_classes>2 else one_hot_bin(y)
    cnt += X.shape[0]
del X, y  
```

```python
Emb_Random= Emb_Random.numpy()
Emb_SSL= Emb_SSL.numpy()
Emb_FT= Emb_FT.numpy()
if nb_classes==1:
    Label_stack= Label_stack.numpy()[...,0]
else:
    Label_stack= torch.argmax(Label_stack, dim=1).numpy() 
```
