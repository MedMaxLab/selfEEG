```python
# IMPORT BASE PACKAGES
import os
import glob
import time
import math
import random
import copy
import importlib 
import sys

# IMPORT CLASSICAL PACKAGES
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score

# IMPORT TORCH
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, Sampler

# IMPORT CUSTOM SELF-SUPERVISED LEARNING FOR EEG LIBRARY
from selfeeg import Augmentation as aug
from selfeeg import Dataload as datald
from selfeeg import ModelZoo as zoo
from selfeeg import SSL as ssl
from selfeeg import losses
```

```python
# ---------------------------
#       TRAINING SPECS
# ---------------------------

# SET DEVICE
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print('Using device:', device)

# SET CPU NUMBER OF WORKERS AND BATCH SIZE
workers= 4
batchsize= 256

# --------------------------
#       FILE LOCATION
# --------------------------
eegpath='/data/zanola/_mat_preprocessed/'
#eegpath='Data_Folder'

# --------------------------
#    PARTITION PARAMETERS
# --------------------------
newSplit = False      # Avoid running the partition number extraction if the partition specs haven't been changed
rmBadRec = True       # Remove records with greater than a threshold (or lower than thresh * -1)
badThresh= 9999          # Threshold for bad eeg removal
freq     = 250        # sampling frequency in [Hz]
overlap  = 0.10       # overlap between partitions
window   = 5          # window length in [seconds]
chans    = 61         # number of channels 

# --------------------------
#       SAMPLER SPECS
# --------------------------
Mode=1
Keep_only_ratio=0.1

# ------------------------------------------
#    PRETAINING TRAIN VAL TEST SLPIT SPEC
# ------------------------------------------
# NOTE: CHECK HOW PARTITION IS PERFORMED WITH help(datald.GetEEGSplitTable)
test_ratio            = None
val_ratio             = 0.1
exclude_data_id       = [1,2,3,4,7,8,9,11,12,13,14,15]
test_data_id          = [5, 10]
val_data_id           = None
val_ratio_on_all_data = False
test_split_mode       = 0
val_split_mode        = 1
split_tolerance       = 0.01
perseverance          = 1000
dataset_info          = None
save                  = True

#test_data_id= [5, 13]
#exclude_data_id= [12, 14, 15]

# EXTRACT FROM PREVIOUS TABLES ALL THE FILES IN TEST SET
# THE FILES WILL BE FURTHER PARTITIONED IN TRAIN, VALIDATION AND TEST SET FOR FINE-TUNING
val_ratio_FT=0.2
test_ratio_FT=0.2
exclude_data_id_FT= None
test_data_id_FT= None
val_data_id_FT=None
val_ratio_on_all_data_FT=True
test_split_mode_FT=1
val_split_mode_FT= 1
save_FT=False
subj_info_key='Group'

is_binary_class=False
nb_classes = 1 if is_binary_class else 4
```

```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# -------------------------------
#  DEF ONE HOT ENCODING FUNCTION
# -------------------------------
# The current problem is a 4 class classification task: healthy, parkinson, alzheimer, dementia EEGs
# Dataset 5    -->    PD: Parkinson         -       CTL: Control
# Dataset 10   -->     A: Alzheimer    F: Dementia    C: Control

def one_hot_ND(label):
    OneHot= torch.zeros(len(label), 4, dtype=torch.long)
    for n, i in enumerate(label):
        if i == 'A':
            OneHot[n,1]=1
        elif i == 'F':
            OneHot[n,2]=1
        elif i == 'PD':
            OneHot[n,3]=1
        else:
            OneHot[n,0]=1
    return OneHot


def one_hot_bin(label):
    OneHot= torch.zeros(len(label), dtype=torch.long)
    for n, i in enumerate(label):
        if i in ['A', 'F', 'PD']:
            OneHot[n]=1
        else:
            OneHot[n]=0
    return torch.unsqueeze(OneHot,1)

def count_classes(dataloader, 
                  nclasses: int=2, 
                  class_name: list[str]=None
                 ):
    class_count=[0]*nclasses
    N = range(nclasses)
    if class_name==None:
        class_name=[i for i in N]
    for X , y in dataloader:
        for i in N:
            try: 
                class_count[i]= class_count[i] + sum(1 for x in y if x in class_name[i])
            except:
                class_count[i]= class_count[i] + sum(1 for x in y if x == class_name[i])
    return class_count
```

```python
# ---------------------------------------------
#               SPLIT DATA
# ---------------------------------------------
if newSplit:
    eeglen= datald.GetEEGPartitionNumber(eegpath, freq, window, overlap, save=save, verbose=True)
else:
    eeglen   = pd.read_csv('EEGPartitionNumber.csv') if os.path.exists('EEGPartitionNumber.csv') else None

if rmBadRec:
    BadEEG  = pd.read_csv('min_max_strangefiles.csv', index_col=[0])
    TooHigh = BadEEG.loc[ BadEEG['max']>badThresh,'preproname'].values.tolist()
    TooLow  = BadEEG.loc[ BadEEG['min']<-badThresh,'preproname'].values.tolist()
    ToRem   = set(TooHigh)
    ToRem   =[item.split('/')[-1] for item in ToRem] 
    eeglen = eeglen[~eeglen['file_name'].isin(ToRem)]

eegsplit= datald.GetEEGSplitTable(partition_table=eeglen,
                                  test_ratio= test_ratio,
                                  val_ratio= val_ratio,
                                  exclude_data_id=exclude_data_id,
                                  test_data_id=test_data_id,
                                  val_data_id=val_data_id,
                                  val_ratio_on_all_data=val_ratio_on_all_data,
                                  test_split_mode=test_split_mode,
                                  val_split_mode= val_split_mode,
                                  split_tolerance=split_tolerance,
                                  perseverance=perseverance,
                                  dataset_info = dataset_info,
                                  EEGpath=eegpath,
                                  save=save)


filesFT= eegsplit.loc[eegsplit['split_set']==2, 'file_name'].values
eeglenFT= eeglen.loc[eeglen['file_name'].isin(filesFT)]
Dataset_length=eeglenFT['N_samples'].sum()
Labels=np.zeros((eeglenFT.shape[0], 1 if is_binary_class else nb_classes))
for i in range(eeglenFT.shape[0]):
    Label = loadmat(eegpath+eeglenFT.iloc[i]['file_name'], simplify_cells=True)
    Label = Label['DATA_STRUCT']['subj_info'][subj_info_key]
    Labels[i,:]= one_hot_bin([Label]) if is_binary_class else one_hot_ND([Label])
Labels_df=pd.DataFrame(Labels, columns=['class_'+str(i) for i in range(nb_classes)])

```

```python
N=100000
M= eeglenFT.shape[0]
count_tot=np.zeros((M,N)).astype('int')
start=time.time()
for i in range(N):
    eegsplitFT = datald.GetEEGSplitTableV2(partition_table        = eeglenFT,
                                         test_ratio             = test_ratio_FT,
                                         val_ratio              = val_ratio_FT,
                                         exclude_data_id        = exclude_data_id_FT,
                                         test_data_id           = test_data_id_FT,
                                         val_data_id            = val_data_id_FT,
                                         val_ratio_on_all_data  = val_ratio_on_all_data_FT,
                                         test_split_mode        = test_split_mode_FT,
                                         val_split_mode         = val_split_mode_FT,
                                         stratified= True,
                                         labels=np.argmax(Labels, axis=1),
                                         split_tolerance        = split_tolerance,
                                         perseverance           = perseverance,
                                         dataset_info           = dataset_info,
                                         EEGpath                = eegpath,
                                         save                   = save_FT
                                        )
    count_tot[:,i]=eegsplitFT['split_set'].values
    

print(time.time() -start)
```

```python
count_tot_df=pd.DataFrame(count_tot, columns=['split_'+str(i) for i in range(count_tot.shape[1])])
Splits = eeglenFT.copy()
Splits.reset_index(inplace=True)
Splits = Splits.join(Labels_df.set_axis(Labels_df.index))
Splits = Splits.join(count_tot_df.set_axis(count_tot_df.index))
Splits = Splits.drop(columns='Unnamed: 0')
Splits
```

```python
trainRat=np.zeros(N)
valRat=np.zeros(N)
testRat=np.zeros(N)

for i in range(N):
    trainRat[i]=Splits.loc[Splits['split_'+str(i)]==0,'N_samples'].sum()/Dataset_length
    valRat[i]=Splits.loc[Splits['split_'+str(i)]==1,'N_samples'].sum()/Dataset_length
    testRat[i]=Splits.loc[Splits['split_'+str(i)]==2,'N_samples'].sum()/Dataset_length

split_ratio_df=pd.DataFrame([trainRat, valRat, testRat]).transpose()
split_ratio_df.rename(columns={0:'Train', 1:'Validation', 2:'Test'}, inplace=True)
sns.boxplot(data=split_ratio_df)
plt.title('sample ratio '+ str(N) + ' splits')
plt.show()
```

```python
FileRat=np.zeros((M,3))

for i in range(M):
    FileRat[i,0]=(Splits.iloc[i].values[-N:]==0).sum()/N
    FileRat[i,1]=(Splits.iloc[i].values[-N:]==1).sum()/N
    FileRat[i,2]=(Splits.iloc[i].values[-N:]==2).sum()/N

split_ratio_df=pd.DataFrame(FileRat, columns= ['Train', 'Validation', 'Test'])
sns.boxplot(data=split_ratio_df)
plt.title('Files ratio '+ str(N) + ' splits')
plt.show()
```

```python
Class_tot=np.zeros((2 if is_binary_class else nb_classes))
if is_binary_class:
    Class_tot[0]=Splits.loc[Splits['class_0']==0,'N_samples'].values.sum()
    Class_tot[1]=Splits.loc[Splits['class_0']==1,'N_samples'].values.sum()
else:
    for i in range(nb_classes):
        Class_tot[i]=Splits.loc[Splits['class_'+str(i)]==1,'N_samples'].values.sum()

trainClassRat=np.zeros((N, 2 if is_binary_class else nb_classes))
valClassRat=np.zeros((N, 2 if is_binary_class else nb_classes))
testClassRat=np.zeros((N, 2 if is_binary_class else nb_classes))

for i in range(N):
    for j in range(trainClassRat.shape[1]):
        if is_binary_class:
            trainClassRat[i,j]=Splits.loc[ (Splits['class_0']==j) & (Splits['split_'+str(i)]==0), 'N_samples'].values.sum()/Class_tot[j]
            valClassRat[i,j]=Splits.loc[ (Splits['class_0']==j) & (Splits['split_'+str(i)]==0), 'N_samples'].values.sum()/Class_tot[j]
            testClassRat[i,j]=Splits.loc[ (Splits['class_0']==j) & (Splits['split_'+str(i)]==0), 'N_samples'].values.sum()/Class_tot[j]
        else:
            trainClassRat[i,j]=Splits.loc[ (Splits['class_'+str(j)]==1) & (Splits['split_'+str(i)]==0), 'N_samples'].values.sum()/Class_tot[j]
            valClassRat[i,j]  =Splits.loc[ (Splits['class_'+str(j)]==1) & (Splits['split_'+str(i)]==1), 'N_samples'].values.sum()/Class_tot[j]
            testClassRat[i,j] =Splits.loc[ (Splits['class_'+str(j)]==1) & (Splits['split_'+str(i)]==2), 'N_samples'].values.sum()/Class_tot[j]


for i in range(3):
    if i==0:
        sns.boxplot(data=pd.DataFrame(trainClassRat, columns=['class_'+str(i) for i in range(trainClassRat.shape[1])]))
        plt.title('class balance in training set ' + str(N) + ' splits')
    elif i==1:
        sns.boxplot(data=pd.DataFrame(valClassRat, columns=['class_'+str(i) for i in range(valClassRat.shape[1])]))
        plt.title('class balance in validation set ' + str(N) + ' splits')
    elif i==2:
        sns.boxplot(data=pd.DataFrame(testClassRat, columns=['class_'+str(i) for i in range(testClassRat.shape[1])]))
        plt.title('class balance in test set ' + str(N) + ' splits')
    plt.show()
```

```python
trainClassRat=np.zeros((N, 2 if is_binary_class else nb_classes))
valClassRat=np.zeros((N, 2 if is_binary_class else nb_classes))
testClassRat=np.zeros((N, 2 if is_binary_class else nb_classes))

for i in range(N):
    TrainTot=Splits.loc[Splits['split_'+str(i)]==0, 'N_samples'].values.sum()
    ValTot=Splits.loc[Splits['split_'+str(i)]==1, 'N_samples'].values.sum()
    TestTot=Splits.loc[Splits['split_'+str(i)]==2, 'N_samples'].values.sum()
    
    for j in range(trainClassRat.shape[1]):
        if is_binary_class:
            trainClassRat[i,j]=Splits.loc[ (Splits['class_0']==j) & (Splits['split_'+str(i)]==0), 'N_samples'].values.sum()/TrainTot
            valClassRat[i,j]=Splits.loc[ (Splits['class_0']==j) & (Splits['split_'+str(i)]==0), 'N_samples'].values.sum()/ValTot
            testClassRat[i,j]=Splits.loc[ (Splits['class_0']==j) & (Splits['split_'+str(i)]==0), 'N_samples'].values.sum()/TestTot
        else:
            trainClassRat[i,j]=Splits.loc[ (Splits['class_'+str(j)]==1) & (Splits['split_'+str(i)]==0), 'N_samples'].values.sum()/TrainTot
            valClassRat[i,j]  =Splits.loc[ (Splits['class_'+str(j)]==1) & (Splits['split_'+str(i)]==1), 'N_samples'].values.sum()/ValTot
            testClassRat[i,j] =Splits.loc[ (Splits['class_'+str(j)]==1) & (Splits['split_'+str(i)]==2), 'N_samples'].values.sum()/TestTot


for i in range(3):
    if i==0:
        sns.boxplot(data=pd.DataFrame(trainClassRat, columns=['class_'+str(i) for i in range(trainClassRat.shape[1])]))
        plt.title('class balance (normalized on split size) in training set ' + str(N) + ' splits')
    elif i==1:
        sns.boxplot(data=pd.DataFrame(valClassRat, columns=['class_'+str(i) for i in range(valClassRat.shape[1])]))
        plt.title('class balance (normalized on split size) in validation set ' + str(N) + ' splits')
    elif i==2:
        sns.boxplot(data=pd.DataFrame(testClassRat, columns=['class_'+str(i) for i in range(testClassRat.shape[1])]))
        plt.title('class balance (normalized on split size) in test set ' + str(N) + ' splits')
    plt.show()
```
